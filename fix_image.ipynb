```python
import math 
n= 20
```

```python
def coprime(m):
    return [i for i in range(2,m) if math.gcd(i,m)==1]
```

```python
def one_generate(root_element, limit_n):
    return {(root_element**i)%limit_n for i in range(1,limit_n)}
```

```python
one_generate(3,20)
```

```python
def add_element(group, one_generate_new_element, limit_n):
    temp_group = group.copy()
    for x in group:
        for new_element in one_generate_new_element:
            z = (x * new_element)%limit_n
            if z not in group:
                #thêm phần tử vào cấu trúc dữ liệu set của python dùng hàm 
                temp_group.add(z)
    return temp_group    
```

```python
def all_close_group(n):
    coprime_elements = coprime(n)
    close_groups = [{1}]
    one_generate_elements = dict((element, one_generate(element, n)) for element in coprime_elements)
    for element in coprime_elements:
        new_groups = []
        for group in close_groups:
            new_group = add_element(group,one_generate_elements[element], n)
            if new_group not in new_groups and new_group not in close_groups:
                new_groups.append(new_group)
        close_groups.extend(new_groups)
    return close_groups


```

```python
n = 9
close_groups = all_close_group(n)
close_groups
```

```python
n =105 
close_groups = all_close_group(n)
close_groups
```

```python
#vẽ biểu đồ n và length của close_groups(n)
import matplotlib.pyplot as plt
n = 105
x = [i for i in range(1,n+1)]
y = [len(all_close_group(i)) for i in range(1,n+1)]
#plot có chú thích
plt.plot(x,y)
```
